```python
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
file = open("lab1_text.txt", "r")
lines = file.readlines()
lines = list(filter(None, ["".join(char for char in line if char.isalpha()).lower() for line in lines]))
pt = ""
for i in lines:
    pt += i
counter = Counter()
length = 0
for line in lines:
    length += len(line)
    counter += Counter(line)
counter
```

```python
def create_dict():
    letters = "а, б, в, г, ґ, д, е, є, ж, з, и, і, ї, й, к, л, м, н, о, п, р, с, т, у, ф, х, ц, ч, ш, щ, ь, ю, я".split(", ")
    dict = {letters[i]:i for i in range(33)}
    return dict

def create_reverse_dict():
    letters = "а, б, в, г, ґ, д, е, є, ж, з, и, і, ї, й, к, л, м, н, о, п, р, с, т, у, ф, х, ц, ч, ш, щ, ь, ю, я".split(", ")
    dict = {i:letters[i] for i in range(33)}
    return dict

def letter_addition(s1, s2, dict):
    res = ""
    for i in range(len(s1)):
        index = (dict[s1[i]] + dict[s2[i]])%33
        for key, value in dict.items():
            if value == index:
                res += key
    return res

def vigenere_cipher(sk, pt, dict):
    key_string = ""
    remainder = len(pt)%len(sk)
    while len(key_string) < len(pt)-remainder:
        key_string += sk
    if len(sk) != len(pt):
        key_string += sk[:remainder]
    print(pt)
    return letter_addition(pt, key_string, dict)

dictionary = create_dict()
#key = "хібаревутьволи"
key = "варениченко"
ct = vigenere_cipher(key, pt, dictionary)
ct
```

```python
def get_nth_letter(string, n):
    new = []
    for i in range(n):
        #print(string)
        new.append(string[::n].strip())
        string = " "*(n-1) + string
    return new

def index_of_coincidence(seq):
    n = len(seq)
    frequencies = Counter(seq)
    return sum([i*(i-1) for i in list(frequencies.values())])/(n*(n-1))

def average_ic(array):
    sum = 0
    for seq in array:
        sum += index_of_coincidence(seq)
    return sum/len(array)

key_length_ic = {}
for i in range(5, 21):
    key_length_ic.update({average_ic(get_nth_letter(ct, i)):i})
print(key_length_ic)
period = key_length_ic.get(max(key_length_ic.keys()))
print(period)
```

```python
def chi_squared(ct, ef):
    frequencies = Counter(ct)
    letters = "а, б, в, г, ґ, д, е, є, ж, з, и, і, ї, й, к, л, м, н, о, п, р, с, т, у, ф, х, ц, ч, ш, щ, ь, ю, я".split(", ")
    for letter in letters:
        if letter not in frequencies.keys():
            frequencies.update({letter:0})
    sum = 0
    for i in letters:
        temp = ef.get(i)*len(ct)
        sum += (frequencies.get(i) - temp)**2/temp
    return sum

def caesar_decode(n, seq):
    alphabet = create_dict()
    alphabet_but_in_reverse = create_reverse_dict()
    res = ""
    for i in seq:
        index = (alphabet.get(i) - n)%33
        res += alphabet_but_in_reverse.get(index)
    return res

expected_frequencies = {'а': 0.0834, 'б': 0.0153, 'в': 0.0550, 'г': 0.0159, 'ґ': 0.0001, 'д': 0.0306, 'е': 0.0459, 'є': 0.0039, 'ж': 0.0071, 'з': 0.0210, 'и': 0.06, 'і': 0.0623, 'ї': 0.0084, 'й': 0.0124, 'к': 0.04, 'л': 0.0393, 'м': 0.0302, 'н': 0.0710, 'о': 0.0928, 'п': 0.0284, 'р': 0.0548, 'с': 0.0457, 'т': 0.0477, 'у': 0.0338, 'ф': 0.0035, 'х': 0.0117, 'ц': 0.0102, 'ч': 0.0115, 'ш': 0.0071, 'щ': 0.0032, 'ь': 0.0183, 'ю': 0.007, 'я': 0.0216}

reverse_dict = create_reverse_dict()
array = get_nth_letter(ct, period)
restored_key = ""
for sequence in array:
    chis = {}
    for i in range(33):
        temp = caesar_decode(i, sequence)
        chis.update({chi_squared(temp, expected_frequencies):i})
    print(chis.get(min(chis.keys())), min(chis.keys()), reverse_dict.get(chis.get(min(chis.keys()))))
    restored_key += reverse_dict.get(chis.get(min(chis.keys())))
print("restored key:", restored_key)
```
